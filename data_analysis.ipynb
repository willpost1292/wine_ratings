```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.read_csv('dataset/winemag-data-130k-v2.csv')
```

```python
score_distribution_df = ds.points.value_counts().reset_index()
plt.figure(figsize=(16,8))
ax = sns.barplot(x = 'index', y = 'points', data = score_distribution)
ax.set(title="Number of Reviews by Score", xlabel='Score', ylabel='Number of Reviews')
plt.show()
```

```python
ds['title'].str.extractall(r'.*(\d{4}).*')
```

```python
ds['vintage'] = ds['title'].str.extract(r'.*(19[0-9]{2}|20[01][0-9]|2020).*')
ds.vintage = ds.vintage.astype('string')
```

```python
vintage_df = ds.vintage.value_counts().reset_index()
vintage_df = vintage_df[~(vintage_df.vintage < 10)]
```

```python
plt.figure(figsize=(16,8))
ax = sns.barplot(x = 'index', y = 'vintage', data = vintage_df)
ax.set(title="Number of Reviews by Vintage", xlabel='vintage', ylabel='Number of Reviews')
plt.show()
```
